```python
from pathlib import Path
import pandas as pd
import numpy as np

mpg=pd.read_csv(Path("auto_mpg_notnull.csv"))
```

```python
mpg.head()
```

```python
mpg = mpg.drop(['car_name'], axis=1, inplace=False)
mpg.head()
```

```python
mpg['origin'].replace({1:'USA', 2:'EU', 3:'JPN'},inplace=True)
mpg.head()
```

```python
mpg.info()
```

```python
mpg['horsepower'] = pd.to_numeric(mpg['horsepower'], errors='coerce')
mpg['horsepower'].dtypes
```

```python
mpg.describe()
```

```python
mpg_label = mpg['mpg'] 
mpg_feature = mpg.drop(['mpg','origin'], axis=1, inplace=False) 
```

```python
from sklearn.model_selection import train_test_split

Y = mpg_label
X = mpg_feature

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2)
x_train
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression() 
lr.fit(x_train,y_train)
```

```python
mpg_predict = lr.predict(x_test)
mpg_predict[:5]
```

```python
y_test.iloc[:5]
```

```python
from sklearn.metrics import mean_squared_error

lin_rmse = mean_squared_error(y_test, mpg_predict,
                              squared=False)
lin_rmse
```
