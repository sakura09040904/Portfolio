```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn import metrics
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

```python
df.info()
```

```python
df['output'].value_counts()
```

```python
plt.hist(df['trtbps'])
plt.xlabel('Resting Blood Pressure')
plt.ylabel('Value')
plt.show()
```

```python
data_to_boxplot = [df['age'], df['trtbps'], df['chol'], df['thalachh'], df['oldpeak']]
plt.boxplot(data_to_boxplot)
plt.xlabel('Attributes')
plt.ylabel('Value')
plt.show()
```

```python
df.isnull().sum()
```

```python
training_points = df.drop(columns=['output'])
training_label = df['output']

X_train, X_test, Y_train, Y_test = train_test_split(training_points, training_label, test_size=0.3, random_state=4)
X_train.shape
```

```python
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train, Y_train)
guesses = classifier.predict(X_test)

confusion_matrix(Y_test, guesses)
```

```python
metrics.accuracy_score(Y_test,guesses)
```

```python
k_range = range(1,50)

accuracy_score = []

for k in k_range:
    classifier = KNeighborsClassifier(n_neighbors=k)
    classifier.fit(X_train, Y_train)
    guesses = classifier.predict(X_test)
    accuracy_score.append(metrics.accuracy_score(Y_test,guesses))
accuracy_score
```

```python
plt.plot(k_range, accuracy_score)
plt.xlabel('value of KNN')
plt.ylabel('test of accuracy')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler 

df_model = df.copy()
scaler = StandardScaler()
features = [['age', 'trtbps', 'chol', 'thalachh', 'oldpeak']] 

for feature in features:
    df_model[feature] = scaler.fit_transform(df_model[feature]) 
    training_points = df_model.drop(columns=['output'])
    training_labels = df_model['output']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(training_points, training_label, test_size=0.2, random_state=4)

k_range = range(1,50)

accuracy_score = []

for k in k_range:
    classifier = KNeighborsClassifier(n_neighbors=k)
    classifier.fit(X_train, Y_train)
    guesses = classifier.predict(X_test)
    accuracy_score.append(metrics.accuracy_score(Y_test,guesses))
```

```python
plt.plot(k_range, accuracy_score)
plt.xlabel('value of KNN')
plt.ylabel('test of accuracy')
plt.show()
```
