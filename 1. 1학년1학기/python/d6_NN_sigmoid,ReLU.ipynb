```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x=np.arange(-5.0,5.0,0.1)
y=sigmoid(x)
plt.plot(x,y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
def relu(x):
    return np.maximum(0,x)
```

```python
x=np.arange(-5.0,5.0,0.1)
y=relu(x)
plt.plot(x,y)
plt.xlim(-6.0, 6.0)
plt.ylim(-1.0, 5.5)
plt.show()
```
